# Gold price from 1978 to 2020

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gold_price = pd.read_excel('data/Gold_Prices.xlsx')
```

```python
gold_price.head(5)
```

```python
gold_price.columns
```

```python
# Exclusão de colunas desnecessárias
gold_price.drop(['US dollar.1','Euro.1','US dollar.2','Chinese renmimbi.1','Canadian dollar.1'], axis=1, inplace=True)
```

```python
print(gold_price.shape)
gold_price.duplicated() # Verifica se tem dados duplicados
gold_price = gold_price.drop_duplicates() # Remove possíveis dados duplicados
print(gold_price.shape)
```

```python
# Atribui a mediana aos valores nulos
gold_price.fillna({'US dollar':      float( '%.2f'%(gold_price['US dollar'].median()) ),       
                   'Euro':           float( '%.2f'%(gold_price['Euro'].median()) ),
                   'Japanese yen':   float( '%.2f'%(gold_price['Japanese yen'].median()) ), 
                   'Pound sterling': float( '%.2f'%(gold_price['Pound sterling'].median()) ),
                   'Canadian dollar':float( '%.2f'%(gold_price['Canadian dollar'].median()) ),
                   'Swiss franc':    float( '%.2f'%(gold_price['Swiss franc'].median()) ),
                   'Indian rupee':   float( '%.2f'%(gold_price['Indian rupee'].median()) ),
                   'Chinese renmimbi':float( '%.2f'%(gold_price['Chinese renmimbi'].median()) ),
                   'Turkish lira':   float( '%.2f'%(gold_price['Turkish lira'].median()) ),
                   'Saudi riyal':    float( '%.2f'%(gold_price['Saudi riyal'].median()) ),
                   'Indonesian rupiah':float( '%.2f'%(gold_price['Indonesian rupiah'].median()) ),
                   'UAE dirham':     float( '%.2f'%(gold_price['UAE dirham'].median()) ),
                   'Thai baht':      float( '%.2f'%(gold_price['Thai baht'].median()) ),
                   'Vietnamese dong':float( '%.2f'%(gold_price['Vietnamese dong'].median()) ),
                   'Egyptian pound': float( '%.2f'%(gold_price['Egyptian pound'].median()) ),
                   'Korean won':     float( '%.2f'%(gold_price['Korean won'].median()) ),
                   'Russian ruble':  float( '%.2f'%(gold_price['Russian ruble'].median()) ),
                   'South African rand': float( '%.2f'%(gold_price['South African rand'].median()) ),
                   'Australian dollar':  float( '%.2f'%(gold_price['Australian dollar'].median()) ),}, inplace=True)
```

```python
# EXIBE A EVOLUÇÃO DOS CASOS EM RELAÇÃO AS SEMANAS OBSERVADAS
plt.figure(figsize=(16,7))
line_prices = sns.lineplot(err_style='bars', data = gold_price)
plt.title('Gold price from 1978 to 2020',fontsize=14)
line_prices.set(xlabel='Semanas epidemiológica', ylabel='Qtde sint. reportados')
plt.show()
```
